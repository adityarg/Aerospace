```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
l=np.array([0,0.2493,0.4987,0.7479,0.9949,1.2419])
s=np.array([-0.000054, 0.048197, 0.096662, 0.145027, 0.192837, 0.240622])

m,c=np.polyfit(l,s,1)
print('Best fit line equation: y = ',m,'x ',c)

fig=plt.figure(figsize=(4,3))
plt.plot(l, s, 'o')
plt.plot(l, m*l + c)
plt.xlabel('Load (in Kg)')
plt.ylabel('S2 (in mV)')
fig.savefig('plot_s2.png', format='png', dpi=300)
```
